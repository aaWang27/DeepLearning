```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import tensorflow as tf
from tensorflow import keras
```

```python
df = pd.read_csv('telco_customer_churn.csv')
df.head()
```

```python
df.drop('customerID', axis=1, inplace=True)
```

```python
df.dtypes
```

```python
df.TotalCharges.values
```

```python
df.isnull().sum()
```

```python
pd.to_numeric(df['TotalCharges'], errors='coerce').isnull()
```

```python
df[pd.to_numeric(df['TotalCharges'], errors='coerce').isnull()]
```

```python
df.shape
```

```python
df1 = df[df['TotalCharges']!=' ']
df1.shape
```

```python
pd.to_numeric(df1['TotalCharges'])
```

```python
df1['TotalCharges'] = pd.to_numeric(df1['TotalCharges'])
df1.TotalCharges.dtypes
```

```python
df1[df1.Churn=='No']
```

```python
def visualize_data(feature):
    churn_no = df1[df1.Churn=='No'][feature]
    churn_yes = df1[df1.Churn=='Yes'][feature]
    
    plt.hist([churn_yes, churn_no], color=['green', 'red'], label=['Churn=Yes', 'Churn=No'])
    plt.title('Customer Churn')
    plt.xlabel(feature)
    plt.ylabel('Number of Customers')
    plt.legend()
```

```python
visualize_data('tenure')
```

```python
visualize_data('MonthlyCharges')
```

```python
def print_column_values(df):
    for column in df:
        print('{}: {}'.format(column, df[column].unique()))
```

```python
print_column_values(df1)
```

```python
df1.replace('No internet service', 'No', inplace=True)
```

```python
print_column_values(df1)
```

```python
df1.replace('No phone service', 'No', inplace=True)
print_column_values(df1)
```

```python
yes_no_columns = ['Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'OnlineSecurity', 'OnlineBackup', 
                  'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling', 'Churn']

for col in yes_no_columns:
    df1[col].replace({'Yes': 1, "No": 0}, inplace=True)
```

```python
print_column_values(df1)
```

```python
df1['gender'].replace({'Female': 1, "Male": 0}, inplace=True)
```

```python
df1['gender'].unique()
```

```python
df1.dtypes
```

```python
df2 = pd.get_dummies(data=df1, columns=['InternetService', 'Contract', 'PaymentMethod'])
df2.columns
```

```python
df2.sample(3)
```

```python
df2.dtypes
```

```python
cols_to_scale = ['tenure', 'MonthlyCharges', 'TotalCharges']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
```

```python
df2.sample(3)
```

```python
print_column_values(df2)
```

```python
X = df2.drop('Churn', axis=1)
y = df2['Churn']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=31)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train[:10]
```

```python
len(X_train.columns)
```

```python
model = keras.Sequential([
    keras.layers.Dense(20, input_shape=(26,), activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100)
```

```python
model.evaluate(X_test, y_test)
```

```python
y_hat = model.predict(X_test)
y_hat[:5]
```

```python
y_test[:5]
```

```python
y_pred = []
for element in y_hat:
    if element > 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
y_pred[:10]
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
import seaborn as sn
plt.figure(figsize=(12,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
